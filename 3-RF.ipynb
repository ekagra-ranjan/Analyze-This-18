```python
#scoring funtion
def scoring(yhat0, y):
    
    y = np.array(y)
    y.sort()
    
    pointsMax = 0
    thresh=0
    
#     print("yhat0 type", type(yhat0))
#     print("yhat0 shape", type(yhat0.shape))
    
#     print("y type", type(y))
#     print("y shape", type(y.shape))
    
    for thresh in np.arange(0.1, 0.9, 0.1):
        points = 0
        budget = 50000
        yhat = np.array(yhat0>thresh, dtype=yhat0.dtype)
        for i in range(len(y)):
            if y[i] == 0:
                if yhat[i] == 0:
                    points+=100
                    budget-=5
                else:
                    budget-=10
            else:
                if yhat[i]==0:
                    points-=50
                    budget-=10
                else:
                    points+=100
                    budget-=10
            if budget<=0:
                break
        if pointsMax<points:
            pointsMax=points
            threshMax = thresh
            
        print(thresh, " ", points)
            
    return pointsMax, threshMax
```
